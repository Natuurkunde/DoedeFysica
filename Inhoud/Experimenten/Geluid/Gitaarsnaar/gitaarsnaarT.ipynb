Bij hoge temperaturen (niet max) 0.621 Ohm, kamertemperatuur: 0.600 Ohm.

```python
import numpy as np
import matplotlib.pyplot as plt

U = np.array([0, 0.303, 0.503, 0.704,0.904,1.105,1.306,1.506,1.707,1.807,2.007,2.207,2.408,2.608,2.808,3.008]) #V
I = np.array([0,0.42,0.683,0.95,1.254,1.515,1.817,2.112,2.388,2.533,2.793,2.858,3.307,3.55,3.775,3.986]) #A
f = np.array([112.0,111.7,110.8,109.8,107.8,106.1,104.0,101.0,99.6,98.1,95.0,94.0,86.0,80.0,77.5,74.8]) #Hz

P = U * I #W

plt.figure(figsize=(3, 3))
plt.plot(U, I, 'o-', label='Current (I) vs Voltage (U)', color='blue')
plt.xlabel('Voltage (U) [V]')
plt.ylabel('Current (I) [A]')
plt.show()

```

```python
plt.figure(figsize=(3, 3))
plt.plot(P**.25,f**2,'k.')
plt.xlabel('temperature measure from P^0.25 [K]')
plt.ylabel('f^2 [Hz]')
plt.show()
```

```python
# Snelle meting 2

U_2 = np.array([0, 1.005, 2.008, 3.009, 3.8]) #V
I_2 = np.array([0, 1.636, 3.100, 4.246, 5.000]) #A
f_2 = np.array([110, 105, 91, 73, 55])

plt.figure(figsize=(3, 3))
plt.plot(U_2, I_2, 'o-', label='Current (I) vs Voltage (U)', color='blue')
plt.xlabel('Voltage (U) [V]')
plt.ylabel('Current (I) [A]')  
plt.show()

plt.figure(figsize=(3, 3))
plt.plot((U_2*I_2), f_2**2, 'k.')
plt.xlabel('Power (U * I) [W^]')
plt.ylabel('f^2 [Hz^2]')
plt.show()
```
