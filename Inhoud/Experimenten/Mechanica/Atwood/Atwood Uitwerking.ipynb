# Uitwerkingen van de metingen van toestel van Atwood

Met de werkformule:

$$(m_1+m_2) \cdot a = m_2 \cdot g$$

De versnelling a werd gemeten voor verschillende waarden van m_2. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Verkregen data
m2 = np.array([
    3.0143, 4.0116, 5.0202, 6.0195, 7.0173, 8.0152, 9.0132,
    10.0120, 11.0303, 12.0292, 12.9503, 13.9493, 14.9469, 15.9551, 16.9529,
    17.9732, 18.9799, 19.9797, 20.9762, 21.9847, 22.9818, 23.9800, 24.9779,
    25.9860, 27.0051, 28.0043, 29.0310, 30.0019])*1e-3  # kg

a = np.array([
    0.134, 0.170, 0.209, 0.266, 0.326, 0.369, 0.421,
    0.455, 0.499, 0.543, 0.5824, 0.6198, 0.6595, 0.6980, 0.7363,
    0.7769, 0.8241, 0.8614, 0.8996, 0.9357, 0.9769, 1.012, 1.050,
    1.087, 1.124, 1.161, 1.198, 1.242]) # m/s^2

print(m2[::3])
print(a[::3])
```

```python
# Model en fit
def atwood_model(m2, m1):
    return (9.81 * m2) / (m2 + m1)

var, cov = curve_fit(atwood_model, m2[::3], a[::3], p0=[200])
m1 = var[0]
m1_err = np.sqrt(np.diag(cov))

print(f"Bepaling m_1 (karretje) = ({m1:.4f} ± {m1_err[0]:.4f}) kg")
```

```python
# Plotten
m2_fit = np.linspace(0.8*min(m2), 1.2*max(m2), 300)
a_fit = atwood_model(m2_fit, *var)

plt.figure(figsize=(10, 6))

plt.xlabel('$m_2$ [kg]')
plt.ylabel('$a$ [m/s²]')

plt.plot(m2[::3], a[::3], 'k.', label='Meetpunten')
plt.plot(m2_fit, a_fit, 'r--', label='Fit: a = g·m₂ / (m₁ + m₂)')

plt.xlim(0.9*min(m2), 1.1*max(m2))
plt.grid(True)
plt.legend()

plt.show()
```

Ook is de massa van het karretje m{sub}`2` gewogen wat (199,3176±0,00005)  g bedraagt. H{sub}`2`O
