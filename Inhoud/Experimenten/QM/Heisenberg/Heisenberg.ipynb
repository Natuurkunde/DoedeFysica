```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit
```

```python
dx = np.array([5,8,11,14,17,20])*1e-4 # slitwidth in m
S = np.array([36.5,46.40,22.35,18.30,13.90,11.40])*1e-3 # spot width in m #68.90


L = 150 # cm
h = 6.63e-34

dp = S * h / (2 * L * 650e-9)

def fit(dx, a):
    return a * dx**(-1)

var, cov = curve_fit(fit, dx, dp )

dx_fit = np.linspace(0.8*min(dx), 1.2*max(dx), 100)
dp_fit = fit(dx_fit, var[0])
```

```python
plt.figure()

plt.plot(1/dx, dp, 'k.')
plt.plot(1/dx_fit, dp_fit, 'r--')

plt.xlabel(r'$\frac{1}{\Delta x}$ (m$^{-1}$)')
plt.ylabel('$\Delta p$ (cm)')

plt.xlim(0, 1.2*max(1/dx))

plt.show()
```
